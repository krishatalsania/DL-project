```python
# =============================================================================
# üëÅÔ∏è GOOGLE COLAB TRAINING SCRIPT - VGG19 (Advanced Fine-Tuning)
# =============================================================================

import os
import zipfile
import shutil
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
from google.colab import files
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix

# ==========================================
# ‚öôÔ∏è CONFIGURATION
# ==========================================
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS_HEAD = 10      # Phase 1: Train classification head
EPOCHS_FINE = 10      # Phase 2: Fine-tune VGG layers
MODEL_NAME = "final_eye_disease_model.h5"
ZIP_NAME = "eye_diseas_detetction.zip"
ZIP_PATH = f"/content/{ZIP_NAME}"
EXTRACT_DIR = "/content/dataset_extracted"
CONFIDENCE_THRESHOLD = 0.75  # 75% confidence required to accept a result

def main():
    # 1. CHECK GPU STATUS
    print(f"TensorFlow Version: {tf.__version__}")
    gpu_info = tf.config.list_physical_devices('GPU')
    if gpu_info:
        print(f"‚úÖ GPU Detected: {gpu_info}")
    else:
        print("‚ö†Ô∏è No GPU detected. Go to Runtime > Change runtime type > Select T4 GPU")

    # 2. EXTRACT DATASET
    if not os.path.exists(EXTRACT_DIR):
        if not os.path.exists(ZIP_PATH):
            print(f"‚ùå Error: '{ZIP_NAME}' not found in /content/!")
            print("üëâ Please drag 'dataset.zip' into the sidebar files.")
            return

        print(f"‚úÖ Found '{ZIP_NAME}'. Extracting...")
        with zipfile.ZipFile(ZIP_PATH, 'r') as zip_ref:
            zip_ref.extractall(EXTRACT_DIR)
        print("‚úÖ Extraction Complete.")

    # Locate data directory
    data_dir = EXTRACT_DIR
    for root, dirs, files_list in os.walk(EXTRACT_DIR):
        lower_dirs = [d.lower() for d in dirs]
        if 'normal' in lower_dirs or 'cataract' in lower_dirs:
            data_dir = root
            break

    print(f"üìç Data Directory: {data_dir}")

    # 3. DATA GENERATORS (With stronger augmentation)
    print("\nüîÑ Initializing Generators...")
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,       # Increased rotation
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.15,
        zoom_range=0.15,
        horizontal_flip=True,
        fill_mode='nearest',
        brightness_range=[0.8, 1.2], # Handle lighting variations
        validation_split=0.2
    )

    train_generator = train_datagen.flow_from_directory(
        data_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )

    val_generator = train_datagen.flow_from_directory(
        data_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='validation',
        shuffle=False
    )

    class_map = {v: k for k, v in train_generator.class_indices.items()}
    num_classes = len(class_map)
    print(f"üè∑Ô∏è Classes: {train_generator.class_indices}")

    # 4. BUILD MODEL (VGG19)
    print("\nüß† Building Model...")
    base_model = VGG19(weights='imagenet', include_top=False, input_shape=IMG_SIZE + (3,))

    # Initially freeze ALL base layers
    base_model.trainable = False

    x = Flatten()(base_model.output)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x) # Helps generalization
    x = Dropout(0.5)(x)         # Higher dropout to reduce overfitting
    output = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=base_model.input, outputs=output)

    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # 5. PHASE 1: TRAIN HEAD
    print("\nüöÄ Phase 1: Training Classification Head...")
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=4, verbose=1, restore_best_weights=True),
        ModelCheckpoint(MODEL_NAME, monitor='val_accuracy', save_best_only=True, verbose=1)
    ]

    history1 = model.fit(
        train_generator,
        validation_data=val_generator,
        epochs=EPOCHS_HEAD,
        callbacks=callbacks
    )

    # 6. PHASE 2: FINE TUNING (Crucial for Real World Accuracy)
    print("\nüîì Phase 2: Unfreezing VGG19 for Fine-Tuning...")
    base_model.trainable = True

    # Freeze the first 15 layers, unfreeze the rest (Block 5)
    # VGG19 has 22 layers (including pools). We train the last conv block.
    for layer in base_model.layers[:17]:
        layer.trainable = False

    # Recompile with VERY LOW learning rate
    model.compile(optimizer=Adam(learning_rate=1e-5),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    print("üöÄ Starting Fine-Tuning...")
    history2 = model.fit(
        train_generator,
        validation_data=val_generator,
        epochs=EPOCHS_FINE,
        callbacks=callbacks
    )

    # 7. SAVE & DOWNLOAD
    print("\nüíæ Saving Final Model...")
    model.save(MODEL_NAME)

    plot_history(history1, history2)
    evaluate_model(model, val_generator)

    # 8. TEST FUNCTION WITH THRESHOLD
    print("\nüß™ Testing Threshold Logic...")
    # This demonstrates how to handle 'Unknown' images
    test_prediction_logic(model, class_map)

    print("\n‚¨áÔ∏è Downloading files...")
    try:
        files.download(MODEL_NAME)
        files.download('training_plots.png')
        files.download('confusion_matrix.png')
    except:
        print("‚ö†Ô∏è Manual download required.")

def test_prediction_logic(model, class_map):
    """
    Demonstrates the logic you should use in your API/Frontend
    to handle unknown images.
    """
    print("-" * 40)
    print(f"‚ÑπÔ∏è LOGIC FOR FRONTEND: Confidence Threshold = {CONFIDENCE_THRESHOLD*100}%")
    print("-" * 40)

    # Mocking a prediction to show logic
    # In real use, you pass an image array here
    # 1. High confidence example
    mock_pred_high = np.array([[0.05, 0.90, 0.02, 0.03]]) # 90% class 1
    process_prediction(mock_pred_high, class_map, "High Confidence Example")

    # 2. Low confidence example (e.g. random image)
    mock_pred_low = np.array([[0.30, 0.30, 0.20, 0.20]]) # Confused model
    process_prediction(mock_pred_low, class_map, "Low Confidence/Unknown Example")

def process_prediction(pred_array, class_map, description):
    idx = np.argmax(pred_array[0])
    confidence = pred_array[0][idx]

    print(f"\nScenario: {description}")
    print(f"Raw Output: {pred_array}")

    if confidence < CONFIDENCE_THRESHOLD:
        print(f"üî¥ Result: NO DISEASES FOUND / UNKNOWN (Confidence {confidence:.2f} < {CONFIDENCE_THRESHOLD})")
        print("   -> Since no single class reached the threshold, we assume the input is invalid or healthy/unknown.")
    else:
        label = class_map[idx]
        print(f"üü¢ Result: {label} (Confidence: {confidence:.2f})")

def plot_history(h1, h2):
    # Combine histories
    acc = h1.history['accuracy'] + h2.history['accuracy']
    val_acc = h1.history['val_accuracy'] + h2.history['val_accuracy']
    loss = h1.history['loss'] + h2.history['loss']
    val_loss = h1.history['val_loss'] + h2.history['val_loss']

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(acc, label='Train')
    plt.plot(val_acc, label='Val')
    plt.axvline(x=len(h1.history['accuracy']), color='green', linestyle='--', label='Fine Tuning Start')
    plt.title('Accuracy (Phase 1 + 2)')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(loss, label='Train')
    plt.plot(val_loss, label='Val')
    plt.axvline(x=len(h1.history['loss']), color='green', linestyle='--', label='Fine Tuning Start')
    plt.title('Loss (Phase 1 + 2)')
    plt.savefig('training_plots.png')

def evaluate_model(model, val_generator):
    val_generator.reset()
    preds = model.predict(val_generator)
    y_pred = np.argmax(preds, axis=1)
    y_true = val_generator.classes

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.savefig('confusion_matrix.png')

    print(classification_report(y_true, y_pred))

if __name__ == "__main__":
    main()
```
